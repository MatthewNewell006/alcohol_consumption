### Standard Deviation and Mean of Temp for First two Quarters of the Year 2015

```python
q1_mean = t_comp['FEB'].mean()
q2_mean = t_comp['MAY'].mean()

q1_std = t_comp['FEB'].std()
q2_std = t_comp['MAY'].std()

# total_drink_std = ac_df['number_of_drinks_total'].std()
# annual_temp_std = t_ann_comp['ANN'].std()
```

### 1) Null Hypothesis: (H0) - "People consume more alcohol in Colder weather than they do in Warmer weather."

### 2) Alternative Hypothesis: (H1) - "People consume more alcohol in Warmer weather than Colder Weather."

### Z Test

```python
# ttest,pval = stats.ttest_ind(t_comp.dropna()['FEB'],t_comp.dropna()['MAY'])
# print(ttest, pval)

z_stat1, z_stat2 = (t_comp['FEB'] - q1_mean) / q1_std , ((t_comp['MAY']) - q2_mean) / q2_std
z_stat1, z_stat2
```

### Anderson Darling Test

```python
# data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data = t_comp['FEB']
result = anderson(data)
print('stat=%.3f' % (result.statistic))
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
        print('Probably Gaussian at the %.1f%% level' % (sl))
    else:
        print('Probably not Gaussian at the %.1f%% level' % (sl))
```

### Pearson Correlation r

```python
# data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
# data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
data1 = t_comp['FEB']
data2 = t_comp['NOV']
stat, p = stats.pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('Probably independent')
else:
    print('Probably dependent','/', 'Positive Correlation')
```
